```python
import requests
import matplotlib.pyplot as plt
```

```python
# Send a GET request to retrieve the data from the API
url = "https://data.nasa.gov/resource/y77d-th95.json"
response = requests.get(url)
data = response.json()
```

```python
# Get all the Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = [meteorite for meteorite in data if "year" in meteorite and
                                int(meteorite["year"][:4]) < 2000]

# Print the results
print("Earth meteorites that fell before the year 2000:")
for meteorite in earth_meteorites_before_2000:
    print(meteorite["name"])
```

```python
# Get all the Earth meteorite coordinates that fell before the year 1970
earth_meteorites_before_1970 = [(float(meteorite["reclat"]), float(meteorite["reclong"])) for meteorite in data if
                                "reclat" in meteorite and "reclong" in meteorite and "year" in meteorite and
                                int(meteorite["year"][:4]) < 1970]

# Print the results
print("\nEarth meteorite coordinates that fell before the year 1970:")
for coordinates in earth_meteorites_before_1970:
    print(f"Latitude: {coordinates[0]}, Longitude: {coordinates[1]}")
```

```python
# Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg
earth_meteorites_greater_than_10000kg = [meteorite for meteorite in data if "mass" in meteorite and meteorite["mass"] is not None and
                                         float(meteorite["mass"]) > 10000]

print("\nEarth meteorites with mass greater than 10000kg:")
for meteorite in earth_meteorites_greater_than_10000kg:
    print(meteorite["name"])
```

```python
# Plot the distribution of earth meteorite masses
meteorite_masses = [float(meteorite["mass"]) for meteorite in data if "mass" in meteorite and meteorite["mass"] is not None]
plt.hist(meteorite_masses, bins=5)
plt.xlabel("Meteorite Mass (g)")
plt.ylabel("Frequency")
plt.title("Distribution of Earth Meteorite Masses")
plt.show()
```
